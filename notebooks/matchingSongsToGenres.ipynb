```python
import os
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
data_path = 'data'
genres_path = 'genresAssigned'
genres_filenames = ['msd-MAGD-genreAssignment.cls', 'msd-topMAGD-genreAssignment.cls']

midi_path= os.path.join(data_path, 'lmd_matched')
```

```python
all_matched_midis = set()

for hier1 in os.listdir(midi_path):
    for hier2 in os.listdir(os.path.join(midi_path, hier1)):
        for hier3 in os.listdir(os.path.join(midi_path,hier1,hier2)):
            for actual_name in os.listdir(os.path.join(midi_path, hier1, hier2, hier3)):
                all_matched_midis.add(actual_name)
```

```python
len(all_matched_midis) # How many entries from 1 milion songs were matched
```

```python
midi_genre_map = set()

for f_name in genres_filenames:
    with open(os.path.join(data_path, genres_path, f_name),'r') as f:
        for l in f.readlines():
            try:
                name, genre = l.split()
            except:
                print(l)
            if(name in all_matched_midis):
                midi_genre_map.add((name, genre))
```

```python
print(len(midi_genre_map))
```

```python
count_genres = {}

for id, genre in list(midi_genre_map):
    if genre not in count_genres:
        count_genres[genre] = 1
    else:
        count_genres[genre] +=1
```

```python
figure(figsize=(20, 10), dpi=80)
plt.bar(count_genres.keys(), count_genres.values(), color='g')
```
